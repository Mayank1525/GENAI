```python
import nltk
import json
import string

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text = """
My name is Mayank Vishwakarma and I am pursuing bachelor's in Computer Science from PSIT Kanpur"""

corpus = text.lower()
print(corpus)
```

```python
nltk.download('punkt_tab')
tokens = word_tokenize(corpus)
print(tokens)
```

```python
tokens = [word for word in tokens if word.isalpha()]
print(tokens)
```

```python
stop_words = set(stopwords.words('english'))

clean_tokens = [word for word in tokens if word not in stop_words]
print(clean_tokens)
```

```python
vocabulary = sorted(set(clean_tokens))
print(vocabulary)
```

```python
vocab_dict = {"vocabulary":vocabulary}
with open('vocab.json', 'w') as file:
  json.dump(vocab_dict, file, indent=4)
print("vocabulary saved to vocabulary.json")
```
