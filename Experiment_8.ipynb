```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize, sent_tokenize
import string

text = """
The candidate should develop scalable web applications using Python and SQL databases.
He must analyze requirements and design efficient systems.
"""

text = text.lower()
sentences = sent_tokenize(text)

print("Sentences:")
print(sentences)

tokens = []
for sentence in sentences:
    words = word_tokenize(sentence)
    words = [word for word in words if word not in string.punctuation]
    tokens.extend(words)

print("\nTokens:")
print(tokens)

pos_tags = nltk.pos_tag(tokens)

print("\nTokens with POS Tags:")
for word, tag in pos_tags:
    print(f"{word} -> {tag}")

meaningful_tags = [
    'NN', 'NNS', 'NNP', 'NNPS',      # Nouns
    'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ',  # Verbs
    'JJ', 'JJR', 'JJS'               # Adjectives
]

meaningful_words = [word for word, tag in pos_tags if tag in meaningful_tags]

print("\nMeaningful Words (Nouns, Verbs, Adjectives):")
print(meaningful_words)
```
